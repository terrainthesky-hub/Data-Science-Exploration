```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data set: https://www.kaggle.com/nikdavis/steam-store-games
df = pd.read_csv(r'C:\Users\Lesley\Downloads\steam-store-games\steam.csv')



print(df.shape)
df.head(100)
```

```python
df3 = df[(df['price']!=0)]
```

```python
#List comprehension to create a binary column as to whether a game has a non-zero average playtime

df['non_zero_avg_played'] = [0 if x == 0 else 1 for x in df['average_playtime']]
```

```python
df2 = df[df['non_zero_avg_played']== 1]
```

```python
#Positive rating percentage

df2['pos_rating_percentage'] =  df2['positive_ratings'] / (df2['negative_ratings'] + df2['positive_ratings'])
```

```python
df2['game_min_per_dollar_avg'] = df2['average_playtime'] / df2['price'].where(df['price']!=0)
```

```python
#Separating games into genres/categories

rogue_likes1 = df2[df2['steamspy_tags'].str.contains("Rogue-like") | df['steamspy_tags'].str.contains("Rogue-lite")]
puzzle_games1 = df2[df2['steamspy_tags'].str.contains("Puzzle")]
souls_like1 = df[df['steamspy_tags'].str.contains("Souls-like")]
indies1 = df2[df2['steamspy_tags'].str.contains('Indie')]
jrpg1 = df2[df2['steamspy_tags'].str.contains('JRPG')]
strategy1 = df2[df2['steamspy_tags'].str.contains('Strategy')]
f2p1 = df2[df2['steamspy_tags'].str.contains('Free to Play')]
action1 = df2[df2['steamspy_tags'].str.contains('Action')]
metroidvania1 = df2[df2['steamspy_tags'].str.contains('Metroidvania')]
casual1 = df2[df2['steamspy_tags'].str.contains('Casual')]
mmo1 = df2[df2['steamspy_tags'].str.contains('MMO')]
hacknslash1 = df2[df2['steamspy_tags'].str.contains('Slash')]
platformer1 = df2[df2['steamspy_tags'].str.contains('Platformer')]
fps1 = df2[df2['steamspy_tags'].str.contains('FPS')]
```

```python
#average playtime

rogue_likes = round(rogue_likes1['average_playtime'].mean())
puzzle_games = round(puzzle_games1['average_playtime'].mean())
souls_like = round(souls_like1['average_playtime'].mean())
indie = round(indies1['average_playtime'].mean())
jrpg = round(jrpg1['average_playtime'].mean())
strategy = round(strategy1['average_playtime'].mean())
f2p = round(f2p1['average_playtime'].mean())
action = round(action1['average_playtime'].mean())
metroidvania = round(metroidvania1['average_playtime'].mean())
casual = round(casual1['average_playtime'].mean())
mmo = round(mmo1['average_playtime'].mean())
hacknslash = round(hacknslash1['average_playtime'].mean())
platformer = round(platformer1['average_playtime'].mean())
fps = round(fps1['average_playtime'].mean())
```

```python
#turning playtime to hours, creating that as a dataframe

rogue_likes = round((rogue_likes1['average_playtime'].mean())/60)
puzzle_games = round((puzzle_games1['average_playtime'].mean())/60)
souls_like = round((souls_like1['average_playtime'].mean())/60)
indie = round((indies1['average_playtime'].mean())/60)
jrpg = round((jrpg1['average_playtime'].mean())/60)
strategy = round((strategy1['average_playtime'].mean())/60)
f2p = round((f2p1['average_playtime'].mean())/60)
action = round((action1['average_playtime'].mean())/60)
metroidvania = round((metroidvania1['average_playtime'].mean())/60)
casual = round((casual1['average_playtime'].mean())/60)
mmo = round((mmo1['average_playtime'].mean())/60)
hacknslash = round((hacknslash1['average_playtime'].mean())/60)
platformer = round((platformer1['average_playtime'].mean())/60)
fps = round((fps1['average_playtime'].mean())/60)


means = [rogue_likes, puzzle_games, souls_like, indie, jrpg, f2p, strategy, action, metroidvania, casual, mmo, hacknslash, platformer, fps]
x_list = ['rogue likes', 'puzzle games', 'souls like', 'indie', 'JRPG', 'F2P', 'strategy', 'action', 'metroidvania', 'casual', 'MMO', 'hack n slash', 'platformer', 'FPS']


newdf = list(zip(x_list, means))

newdf = pd.DataFrame(newdf, columns=['Games', 'mean_playtime'])

newdf
```

```python
#Positive rating percentage average in genre
rogue_likesrat = rogue_likes1['pos_rating_percentage'].mean()
puzzle_gamesrat = puzzle_games1['pos_rating_percentage'].mean()
indierat = indies1['pos_rating_percentage'].mean()
jrpgrat = jrpg1['pos_rating_percentage'].mean()
strategyrat = strategy1['pos_rating_percentage'].mean()
f2prat = f2p1['pos_rating_percentage'].mean()
actionrat = action1['pos_rating_percentage'].mean()
metroidvaniarat = metroidvania1['pos_rating_percentage'].mean()
casualrat = casual1['pos_rating_percentage'].mean()
mmorat = mmo1['pos_rating_percentage'].mean()
hacknslashrat = hacknslash1['pos_rating_percentage'].mean()
platformerrat = platformer1['pos_rating_percentage'].mean()
fpsrat = fps1['pos_rating_percentage'].mean()

rat = [rogue_likesrat, puzzle_gamesrat, indierat, jrpgrat, strategyrat, f2prat, actionrat, metroidvaniarat,casualrat, mmorat, hacknslashrat, platformerrat, fpsrat]

x_list2 = ['rogue likes', 'puzzle games', 'indie', 'JRPG', 'strategy', 'F2P','action', 'metroidvania', 'casual', 'MMO', 'hack n slash', 'platformer', 'FPS']
rating = list(zip(x_list2, rat))

ratings = pd.DataFrame(rating, columns=['Games', 'average_positive_rating_percentage'])
```

```python
#Standard deviation of time played in genre

rogue_likesstd = round((rogue_likes1['average_playtime'].std(ddof=1))/60)
puzzle_gamesstd = round((puzzle_games1['average_playtime'].std(ddof=1))/60)
souls_likestd = round((souls_like1['average_playtime'].std(ddof=1))/60)
indiestd = round((indies1['average_playtime'].std(ddof=1))/60)
jrpgstd = round((jrpg1['average_playtime'].std(ddof=1))/60)
strategystd = round((strategy1['average_playtime'].std(ddof=1))/60)
f2pstd = round((f2p1['average_playtime'].std(ddof=1))/60)
actionstd = round((action1['average_playtime'].std(ddof=1))/60)
metroidvaniastd = round((metroidvania1['average_playtime'].std(ddof=1))/60)
casualstd = round((casual1['average_playtime'].std(ddof=1))/60)
mmostd = round((mmo1['average_playtime'].std(ddof=1))/60)
hacknslashstd = round((hacknslash1['average_playtime'].std(ddof=1))/60)
platformerstd = round((platformer1['average_playtime'].std(ddof=1))/60)
fpsstd = round((fps1['average_playtime'].std(ddof=1))/60)






```

```python
rogue_likesstd = round(rogue_likesstd)
puzzle_gamesstd = round(puzzle_gamesstd)
souls_likestd = round(souls_likestd)
indiestd = round(indiestd)
jrpgstd = round(jrpgstd)
strategystd = round(strategystd)
f2pstd = round(f2pstd)
actionstd = round(actionstd)
metroidvaniastd = round(metroidvaniastd)
casualstd = round(casualstd)
mmostd = round(mmostd)
hacknslashstd = round(hacknslashstd)
platformerstd = round(platformerstd)
fpsstd = round(fpsstd)

stddevs = [rogue_likesstd, puzzle_gamesstd, souls_likestd, indiestd, jrpgstd, f2pstd, strategystd, actionstd, metroidvaniastd, casualstd, mmostd, hacknslashstd, platformerstd, fpsstd]

stddf = list(zip(x_list, stddevs))

stddf = pd.DataFrame(stddf, columns=['Games', 'std_deviation_playtime'])
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Bar(x=ratings['Games'],
                y=ratings['average_positive_rating_percentage'],
                name='Positive rating percentage',
                marker_color='rgb(10, 147, 201)', text=ratings['average_positive_rating_percentage']
                ))

fig.update_layout(
    title='How critical we are of the genre',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Positive review ratio average',
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=0,
        y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    barmode='group',
    bargap=0.15, # gap between bars of adjacent location coordinates.
    bargroupgap=0.1 # gap between bars of the same location coordinate.
)
fig.update_traces(texttemplate='%{text:.2r}', textposition='outside')
fig.show()
```
